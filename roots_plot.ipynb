```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Complex roots of
$$ax^2 + bx + c,$$
where
$$b^2 - 4ac < 0$$
and $$a, b, c \in [-10, 10] \cap \mathbb{Z}$$

```python
low = -14
high = -low + 1

roots = []
for a in range(low, high):
    for b in range(low, high):
        for c in range(low, high):
            if (b**2) - (4*a*c) < 0:
                roots.append(np.roots([a,b,c]))

reals = np.array([item for sublist in [root.real for root in roots] for item in sublist])
imags = np.array([item for sublist in [root.imag for root in roots] for item in sublist])
print(len(reals))
```

```python
fig = plt.figure(figsize=(10, 10), facecolor='black')
plt.xlim(-3.5, 3.5)
plt.ylim(-3.5, 3.5)
plt.axis('off')
plt.scatter(reals, imags, s=0.1, cmap="inferno_r", c=np.sqrt(reals**2 + imags**2))

fig.savefig('output/roots.png', dpi=800, facecolor=fig.get_facecolor(), edgecolor='none')
```
